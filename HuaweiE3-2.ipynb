```python
glandTemp=pd.read_csv('./dataset/question3/3-2/GlobalLandTemperatures/GlobalTemperatures.csv')
```

```python
years=[]
for m in range(len(glandTemp['dt'])):
    years.append(int(glandTemp['dt'][m][0:4]))
years=list(set(years))
```

```python
#glandTemp['LandAverageTemperature']
```

```python
g_y_mean_temp=[]
for m in range(266):
    g_y_mean_temp.append(glandTemp['LandAverageTemperature'][m:m+12].mean())
g_y_mean_temp=pd.DataFrame(g_y_mean_temp)
g_y_mean_temp['years']=years
g_y_mean_temp=g_y_mean_temp.rename(columns={0:'global year mean temp'})
g_y_mean_temp=g_y_mean_temp[130:]
```

```python
greyfact=pd.merge(g_y_mean_temp,anomaly_index1)
greyfact=pd.merge(greyfact,anomaly_index2)
greyfact=pd.merge(greyfact,anomaly_index3)
greyfact=pd.merge(greyfact,anomaly_index4)
greyfact
```

```python
greyfact.to_csv('./dataset/greyfact.csv')
```

```python
del greyfact['years']
greyfact=(greyfact - greyfact.min()) / (greyfact.max() - greyfact.min())
greyfact
```

```python
greyfact.to_csv('./dataset/greyfact.csv')
```

```python
std=greyfact.iloc[:,0]
std
```

```python
ce=greyfact.iloc[:,1:]
ce
```

```python
n,m=ce.shape[0],ce.shape[1]
n,m
```

```python
a=zeros([m,n])
for i in range(m):
    for j in range(n):
        a[i,j]=abs(ce.iloc[j,i]-std[j])
        
```

```python
c=amax(a)
d=amin(a)
c,d
```

```python
result=zeros([m,n])
for i in range(m):
    for j in range(n):
        result[i,j]=(d+0.5*c)/(a[i,j]+0.5*c)
```

```python
result=pd.DataFrame([mean(result[i, :]) for i in range(m)])
result
```

```python
result=result.rename(columns={0:'temp'},index={0:'24N-90N'})
result=result.rename(index={1:'44S-24S',2:'90S-64S',3:'24S-EQU'})
result
```
