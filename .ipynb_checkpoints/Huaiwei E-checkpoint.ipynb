```python
import os
import pandas as pd
import numpy as np
#pd.options.display.max_rows=10
import matplotlib.pyplot as plt
```

```python
fileList=os.listdir('./dataset/Canada')
```

```python
for f in fileList:
    stemp=os.listdir('./dataset/Canada/'+f)
    temp=[]
    for s in stemp:
        temp.append(pd.read_csv('./dataset/Canada/'+ f +'/'+s,header=None,sep=',',encoding='utf-8',skiprows=list(i for i in range(25)),engine='python'))
```

```python
for i in temp:
    grouped=temp[i].groupby(temp[i][2])
    grouped.mean()
```

```python
plt.plot(grouped.mean()[9])
```
