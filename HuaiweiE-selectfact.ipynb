```python
NO2=pd.read_csv('./dataset/question2/N2O/v432_N2O_1970_2012.csv',encoding='utf-8',engine='python',skiprows=[i for i in range(7)])
```

```python
CH4=pd.read_csv('./dataset/question2/CH4/v432_CH4_1970_2012.csv',encoding='utf-8',engine='python',skiprows=[i for i in range(7)])
```

```python
g_m_land_temp=pd.read_csv('./dataset/question2/GlobalLandTemperatures/GlobalTemperatures.csv',encoding='utf-8',engine='python')
```

```python
land_temp_years=g_m_land_temp['dt']
list(land_temp_years)
temp_years=[]
for i in list(land_temp_years):
    #print(i[0:4])
    temp_years.append(int(i[0:4]))
landtemp_years=set(temp_years)
```

```python
temp=[]
for i in range(263):
    temp.append(g_m_land_temp[i:i+12]['LandAverageTemperature'].mean())
g_y_land_temp=pd.DataFrame(temp,index=landtemp_years)
g_y_land_temp
```

```python
gtempdf=g_y_land_temp[226:-3].rename(columns={0:'temp'})
gtempdf['key']=gtempdf.index
gtempdf
```

```python
CO2.drop(CO2.columns[0:6], axis=1, inplace=True)
CO2_y_g_mean=pd.DataFrame(CO2.mean())
CO2_y_g_mean['key']=CO2_y_g_mean.index
CO2_y_g_mean=CO2_y_g_mean.rename(columns={0:'co2'})
CO2_y_g_mean=CO2_y_g_mean[9:]
CO2_y_g_mean
```

```python
NO2.drop(NO2.columns[0:6], axis=1, inplace=True)
NO2_y_g_mean=pd.DataFrame(NO2.mean())
NO2_y_g_mean['key']=NO2_y_g_mean.index
NO2_y_g_mean=NO2_y_g_mean.rename(columns={0:'NO2'})
NO2_y_g_mean=NO2_y_g_mean[9:]
NO2_y_g_mean
```

```python
CH4.drop(CH4.columns[0:6], axis=1, inplace=True)
CH4_y_g_mean=pd.DataFrame(CH4.mean())
CH4_y_g_mean['key']=CH4_y_g_mean.index
CH4_y_g_mean=CH4_y_g_mean.rename(columns={0:'CH4'})
CH4_y_g_mean=CH4_y_g_mean[9:]
CH4_y_g_mean
```

```python
selectFactor=pd.merge(CO2_y_g_mean,NO2_y_g_mean,on='key')
selectFactor=pd.merge(selectFactor,CH4_y_g_mean,on='key')
selectFactor
```
